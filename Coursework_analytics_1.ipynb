```python
df_all_pr = df_all[(df_all['is_homework']==True)&(df_all['status']=='done')]
df_all_pr['start_date'] = df_all_pr['start_date'].map(lambda x: x.strftime('%Y-%m'))
df_all_pr['finish_date'] = df_all_pr['finish_date'].map(lambda x: x.strftime('%Y-%m'))
df_all_pr['finish_date'] = df_all_pr['finish_date'].combine_first(df_all_pr['start_date'])

df_all_pr = df_all_pr.groupby('title').apply(lambda course: course.apply(
    lambda x: pd.Series(1, index=pd.date_range(
        x['start_date'], x['finish_date'], closed=None, freq='MS')), axis=1).sum(
    axis=0)).reset_index(level=0).pivot(columns='title', values=0).reindex(
    pd.date_range(df_all_pr.start_date.min(), df_all_pr.finish_date.max(), freq='MS'))
df_all_pr.fillna(0, inplace=True)
df_all_pr = df_all_pr.astype(int)
df_all_pr
```

```python
df_all_tm = df_all[(df_all['lesson_title'].str.contains('Домашняя работа'))]
df_all_tm = df_all_tm.dropna()
```

```python
df_all_tm = df_all_tm.groupby(['title','module_title','student_id']).agg({'start_date':'min', 'finish_date':'max'}).reset_index()
df_all_tm['start_date'] = df_all_tm['start_date'].dt.date
df_all_tm['finish_date'] = df_all_tm['finish_date'].dt.date
df_all_tm['time'] = (df_all_tm['finish_date']-df_all_tm['start_date'])/timedelta(days=1)
df_all_tm = df_all_tm.groupby(['title','module_title'])['time'].agg(['min','max','mean','median']).reset_index()
df_all_tm
```

```python
fig,axes = plt.subplots(15,1, sharex=False, figsize=(15,120))
fig.subplots_adjust(hspace=1.5)

for ax,q in zip(axes.ravel(), df_all_tm.title.unique()):
    tmp_df = df_all_tm[df_all_tm.title.eq(q)]
    tmp_df.plot.line(x='module_title', y='median', ax=ax, fontsize=10, legend=None)
    ax.set_title(q, fontsize=16)
    ax.set_xlabel('Модули', fontsize=12)
    ax.set_ylabel('Медианное время прохождения модуля', fontsize=12)
    ax.tick_params(axis='x',rotation=270)
```

```python
df_all_m = df_all[(df_all['lesson_title'].str.contains('Домашняя работа'))]
df_all_m = df_all_m.groupby(['title','student_id']).agg({'start_date':'min', 'finish_date':'max'}).reset_index()
df_all_m['finish_date'] = df_all_m['finish_date'].fillna(df_all_m.groupby('title')['finish_date'].transform('max'))
df_all_m['time'] = (df_all_m['finish_date']-df_all_m['start_date'])/timedelta(days=1)
df_all_m['month'] = df_all_m['finish_date'].dt.month
df_all_m = pd.pivot_table(df_all_m, values='time', index='month',columns='title',aggfunc=pd.Series.median,fill_value=0)
print('Медианное время прохождения курсов по месяцам')
df_all_m
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=df_all_m)
plt.title('График медианного прохождения курсов по месяцам', fontsize=18)
plt.xlabel('Месяцы',fontsize=14)
plt.ylabel('Время', fontsize=14)
plt.show()
```
